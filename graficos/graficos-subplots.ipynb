# Importando bibliotecas

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

# Gerando valores aleatórios para o eixo Y

```python
valor_minimo = 1000
valor_maximo = 5000
tamanho_amostra = 100

dados = {
            'A': np.random.randint(low=valor_minimo, high=valor_maximo, size=tamanho_amostra),
            'B': np.random.randint(low=valor_minimo, high=valor_maximo, size=tamanho_amostra)
        }

dados
```

# Gerando valores aleatórios para o eixo Y

```python
data_inicio = '1980-01-01'

time = pd.date_range(start=data_inicio, periods=tamanho_amostra, freq = 'Y')

time
```

# Criando o DataFrame

```python
df = pd.DataFrame(dados, index = time)

print('Quantidade de linhas:', df.shape[0])
print('Quantidade de colunas:', df.shape[1])

# Exibindo as primeiras 5 linhas do dataFrame
df.head()
```

# Criando Gráfico

```python
fig = make_subplots(rows = 2, cols = 1)

fig.add_trace(go.Scatter(x = df.index,
                         y = df['A'],
                         mode = 'lines',
                         name = 'Série 1',
                         line = dict(color = 'green')
                        ), row = 1, col = 1)

fig.add_trace(go.Scatter(x = df.index,
                         y = df['B'],
                         mode = 'lines',
                         name = 'Série 2',
                         line = dict(color = 'red')
                        ), row = 2, col = 1)

fig.update_layout(title = 'Séries temporais')

fig.show()
```

```python
# Podemos também plotar ambos os gráficos na mesma linha
fig = make_subplots(rows = 2, cols = 1)

fig.add_trace(go.Scatter(x = df.index,
                         y = df['A'],
                         mode = 'lines',
                         name = 'Série 1',
                         line = dict(color = 'green')
                        ), row = 1, col = 1)

fig.add_trace(go.Scatter(x = df.index,
                         y = df['B'],
                         mode = 'lines',
                         name = 'Série 2',
                         line = dict(color = 'red')
                        ), row = 1, col = 1)

fig.update_layout(title = 'Séries temporais')

fig.show()
```
